```python
import pandas as pd
```

### **Creating a series** 

```python
a= [1,5,9,4,3,7,2,6,3] # Normal List
a = pd.Series(a)
```

```python
a
```

### **Statistics on Series**
- `s.describe()` *Summary stats for Numerical values*
- `s.mean()` *returns mean*
- `s.corr()` *returns the correlation b/w other columns*
- `s.count()` *Returns number of non-null values*
- `s.max()` *Returns largest number*
- `s.min()` *Returns smallest number*
- `s.std()` *returns standard deviation*
- `s.mean()` *returns mean*

```python
a.max()
```

```python
a.count()
```

```python
a.describe()
```

### **Sorting**

```python
a.sort_values()
```

- by defaut everything is soted in ascending order
- for series to be sorted in descending order `a.sort_values(ascending = False)`
- To modify old Series with new sorted Series, type
`a.sort_values(inplace=True)` *This will modify the series by changing its original order with sorted order*

### **Viewing/Inspecting Data**

- `s.head(n)` : First n rows/values
- `s.tail(n)` : Last n rows/values
- `s.shape` : (n,)
- `s.describe()` : Summary Stats for numerical analysis
- `s.value_counts(dropna=False)` : Views unique values and counts

```python
a.shape
```

```python
a.value_counts(dropna=False)
```

### **Data Cleaning**

- `s.fillna(s.mean())`
 **replaces all null values with the mean**
- `s.astype(float)` **converts data type of the series to float**
- `s.replaces(1,'one`) **Replaces all values equal to 1 with 'one'**
- `s.replace([1,3],['one','three'])` **Replaces all 1 with 'one' and 3 with 'three'**
